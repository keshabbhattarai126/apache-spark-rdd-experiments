```python
schema = StructType([
    StructField("sale_id", IntegerType(), True),
    StructField("product", StringType(), True),
    StructField("category", StringType(), True),
    StructField("sale_date", StringType(), True),
    StructField("region", StringType(), True),
    StructField("customer", StringType(), True),
    StructField("qty", IntegerType(), True),
    StructField("unit_price", DoubleType(), True)
])
```

```python
df = spark.createDataFrame(sales_data, schema)
df.show()
df.printSchema()
```

```python
df = df.withColumn("revenue", col("qty") * col("unit_price"))
df.show(5)
```

```python
df.createOrReplaceTempView("sales")
```

```python
spark.sql("SHOW TABLES").show()
```

```python
query1 = """ 
SELECT * FROM sales
"""
result = spark.sql(query1)
result.show()
```

## Filter by category where category = Electronics


```python
query2 = """ SELECT * FROM sales where category = "Electronics" """
result = spark.sql(query2)
result.show()
```

## Group by product, order by revenue

```python
query3 = """
SELECT 
    product,
    COUNT(*) AS orders,
    SUM(qty) AS unit_sold,
    ROUND(SUM(revenue), 2) AS total_revenue
FROM sales
GROUP BY product
ORDER BY total_revenue DESC
"""
result = spark.sql(query3)
result.show()
```

## product, regioin, qty unit_price, revenue, revenue>10000, category = Electronics, revenue DESC

```python
query4 = """
SELECT
    product, region, qty, unit_price, revenue
FROM sales
WHERE revenue>1000 and category ="Electronics" order by revenue DESC
    
"""
result = spark.sql(query4)
result.show()
```

## RANK products by Revenue within each region

```python
query5 = """
select region,product,
    round(sum(revenue),2) as revenue,
    rank() over(
        partition by region
        order by sum(revenue) DESC
        ) as rnk
    from sales
    group by region, product
    order by region, rnk
"""

result = spark.sql(query5)
result.show()
```

```python
## subquery = products over the average revenue

query6 = """
select product, category, total_revenue
from(
select product, category,
    sum(revenue) as total_revenue
from sales
group by product,category
) prod_totals
where total_revenue > (select avg(revenue) from sales)
order by total_revenue desc
"""
result = spark.sql(query6)
result.show()
```
