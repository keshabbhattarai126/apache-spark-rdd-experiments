# Exploratory Data Analysis (EDA) Using PySpark and Spark SQL

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType
```

```python
spark = SparkSession.builder.appName("Data Analysis Supermarket Sales").master("local[*]").getOrCreate()
```

# Loading and Inspection

```python
df_infer = spark.read.csv("supermarket_sales.csv", header=True, inferSchema=True)
df_infer.printSchema()
df_infer.show(10)
```

# Null Value Handling

```python
null_counts = df_infer.select([
    F.sum(F.when(F.col(c).isNull(), 1).otherwise(0)).alias(c)
    for c in df_infer.columns
])
null_counts.show()
```

# Descriptive Statistics

```python
df_infer.select("quantity", "unit_price", "total_sales").describe().show()
```

# EDA Using PySpark DataFrame Methods

```python
# Calcuilate the total Supermarket Revenue
total_revenue_val = df_infer.select(F.sum("total_sales")).collect()[0][0]
print("Total sales:", total_revenue_val)
```

```python
# Find the top 5 highest-selling product categories based on total sales.
top_5_categories = (
    df_infer
    .groupBy("product_category")
    .agg(F.sum("total_sales").alias("total_category_sales"))
    .orderBy(F.desc("total_category_sales"))
    .limit(5)
)

top_5_categories.show()
```

```python
# Determine: - Average quantity sold - Average unit price - Average sales amount
mean_quantity_val = df_infer.select(F.avg("quantity")).collect()[0][0]
mean_unit_price_val = df_infer.select(F.avg("unit_price")).collect()[0][0]
mean_sales_amount_val = df_infer.select(F.avg("total_sales")).collect()[0][0]

print("Calculated mean Quantity:", mean_quantity_val)
print("Calculated mean Unit Price:", mean_unit_price_val)
print("Calculated mean Sales Amoutn:", mean_sales_amount_val)
```

```python
#  Find the city generating the highest revenue.
highest_revenue_city = (
    df_infer
    .groupBy("city")
    .agg(F.sum("total_sales").alias("total_revenue"))
    .orderBy(F.desc("total_revenue"))
    .limit(1)
)

highest_revenue_city.show(truncate=False)
```
